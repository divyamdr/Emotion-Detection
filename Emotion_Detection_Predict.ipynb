```python
from keras.models import load_model
import numpy as np
import cv2
model =load_model('mymodelfinalaifinal.h5')
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from skimage.transform import resize
def detect1(frame):
    try:
        img= resize(frame,(48,48))
        img = np.expand_dims(img,axis=0)
        if(np.max(img)>1):
            img =img/255.0
        prediction =model.predict(img)
        k1 = prediction[0][0]*100
        k2 = prediction[0][1]*100
        k3 = prediction[0][2]*100
        k4 = prediction[0][3]*100
        k5 = prediction[0][4]*100
        k6 = prediction[0][5]*100
        k7 = prediction[0][6]*100
        k1 = round(k1)
        k2 = round(k2)
        k3 = round(k3)
        k4 = round(k4)
        k5 = round(k5)
        k6 = round(k6)
        k7 = round(k7)
        prediction_class = model.predict_classes(img)
        if (prediction_class[0] == 0):
            data =0
        if (prediction_class[0] == 1):
            data =1
        if (prediction_class[0] == 2):
            data =2          
        if (prediction_class[0] == 3):
            data =3
        if (prediction_class[0] == 4):
            data =4
        if (prediction_class[0] == 5):
            data =5 
        if (prediction_class[0] == 6):
            data =6
    except AttributeError:
        print("shape not found")
    return data,k1,k2,k3,k4,k5,k6,k7
```

```python
import numpy as np
from keras.models import load_model
import cv2
from keras.preprocessing.image import img_to_array
#-----------------------------
#opencv initialization

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for (x, y, w, h) in faces:
        print (faces.shape)
        print ("Number of faces detected: " + str(faces.shape[0]))
        print("Data Found")
        cv2.putText(frame, "Number of faces detected: " + str(faces.shape[0]), (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 255), 2)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        roi_color = frame[y:y+h, x:x+w]
        FaceFileName = "face_" + str(y) + ".jpg"
        cv2.imwrite(FaceFileName, roi_color)
        img=cv2.imread(FaceFileName)
        data,k1,k2,k3,k4,k5,k6,k7 = detect1(img)
        if (data== 0):
             cv2.putText(frame,"Angry", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 1):
             cv2.putText(frame,"Disgust ", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 2):
             cv2.putText(frame,"Fear", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 3):
             cv2.putText(frame,"Happy",(x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 4):
             cv2.putText(frame,"Sad", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 5):
             cv2.putText(frame,"Surprise", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        if (data== 6):
             cv2.putText(frame,"Neutral", (x+10,y+h+20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        cv2.putText(frame,"Angry   :"+str(k1)+"%",(x, y-140),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Disgust :"+str(k2)+"%",(x, y-120),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Fear    :"+str(k3)+"%",(x, y-100),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Happy   :"+str(k4)+"%",(x, y-80),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Sad     :"+str(k5)+"%",(x, y-60),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Surprise:"+str(k6)+"%",(x, y-40),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(frame,"Neutral :"+str(k7)+"%",(x, y-20),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
    return frame

# Doing some Face Recognition with the webcam
video_capture = cv2.VideoCapture(0)
while True:
    _,frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray,frame)
    cv2.imshow('Video',canvas)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
