# Step1:Import Keras Libraies

```python

#import keras libraries
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
```

# Step2: Intialize CNN model

```python
model=Sequential()
```

# Step3: Add Convolution Layer

```python
model.add(Conv2D(32,3,3,input_shape=(48,48,3),activation='relu'))#1St parameter =no of features detectors 2nd& 3rd =Size of feature detector, 
#4th input image size,5 th parameter is channel for color=3 gray scale=1,6 th to avoid negative pixels we use activation function
```

# Step4:Add Pooling Layer

```python
model.add(MaxPooling2D(pool_size=(2,2)))#1parmeter=size of pooling matrix
```

# Step5: Add flatten layer

```python
model.add(Conv2D(64,3,3,activation='relu'))#1St parameter =no of features detectors 2nd& 3rd =Size of feature detector, 
model.add(MaxPooling2D(pool_size=(2,2)))#1parmeter=size of pooling matrix
model.add(Conv2D(128,3,3,activation='relu'))#1St parameter =no of features detectors 2nd& 3rd =Size of feature detector, 
model.add(MaxPooling2D(pool_size=(2,2)))#1parmeter=size of pooling matrix
model.add(Conv2D(256,3,3,activation='relu'))#1St parameter =no of features detectors 2nd& 3rd =Size of feature detector, 
model.add(MaxPooling2D(pool_size=(2,2)))#1parmeter=size of pooling matrix

model.add(Flatten())
```

# Step6: Ann starts so need to add dense layers

```python

model.add(Dense(output_dim=128,activation='relu',init='random_uniform'))
```

```python
model.add(Dense(output_dim=7,activation='sigmoid',init='random_uniform'))
```

# Step7:Compile the model

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

# Step8: Preprocess images

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'C:\Users\Lenovo\Desktop\AI_project\train_set',target_size=(48,48),batch_size=8,class_mode='categorical')
x_test = train_datagen.flow_from_directory(r'C:\Users\Lenovo\Desktop\AI_project\test_set',target_size=(48,48),batch_size=8,class_mode='categorical')
```

```python
model.fit_generator(x_train,samples_per_epoch = 1000,epochs=1000,validation_data=x_test,nb_val_samples=110)#(samples_per_epoch= no of traininig or testing images/batch size)
                                                                        #                                                       =8000/32=250
```
